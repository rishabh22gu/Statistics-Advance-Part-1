Q.1- Random Variable in Probability Theory
A random variable is a function that assigns numerical values to outcomes of a random experiment. It can take different values based on the experiment's outcome.

Q.2- Types of Random Variables
Discrete Random Variable: Takes a countable number of distinct values (e.g., the number of heads in coin flips).
Continuous Random Variable: Takes an uncountable number of values within a given range (e.g., a person's height or temperature).

Q.3- Difference Between Discrete and Continuous Distributions
Discrete distributions deal with countable values (e.g., number of defective items in a batch).
Continuous distributions deal with measurable values over an interval (e.g., weights of newborn babies).

Q.4- Probability Distribution Function (PDF)

A probability distribution function describes how probabilities are distributed over the values of a random variable.

For discrete variables, it is called the Probability Mass Function (PMF).
For continuous variables, it is called the Probability Density Function (PDF).

Q.5- Cumulative Distribution Function (CDF) vs. Probability Distribution Function (PDF)

The CDF gives the probability that a random variable is less than or equal to a particular value.
The PDF (for continuous variables) or PMF (for discrete variables) gives the probability of a specific value occurring.

Q.6- Discrete Uniform Distribution

A distribution where all possible values of a discrete random variable are equally likely. Example: Rolling a fair die.

Q.13- Central Limit Theorem (CLT) and Its Importance
States that the sampling distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the original distribution.
Crucial because it allows us to use normal distribution approximations in many statistical analyses.

Q.14- How CLT Relates to the Normal Distribution

Used to test hypotheses when population parameters are known.
Helps in determining p-values and rejecting/accepting null hypotheses.

Q.17- Point Estimates vs. Interval Estimates
Point Estimate: A single value estimate of a parameter (e.g., sample mean).
Interval Estimate: A range within which the parameter is expected to lie (e.g., confidence interval).

Q.18- Significance of Confidence Intervals
Indicates the range in which the true population parameter is likely to be found.
A 95% confidence interval means we are 95% confident that the parameter falls within the interval.

19- Relationship Between Z-Score and Confidence Interval

```python
#2
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define discrete uniform distribution parameters
low, high = 1, 6  # Simulating a fair die roll
x = np.arange(low, high+1)
pmf = stats.randint.pmf(x, low, high+1)

# Plot PMF
plt.bar(x, pmf, alpha=0.7, color='b', edgecolor='black')
plt.xlabel("Values")
plt.ylabel("Probability")
plt.title("Discrete Uniform Distribution PMF")
plt.show()
```

```python
#3
def bernoulli_pdf(p, x):
    return p**x * (1 - p)**(1 - x)

# Example usage
p = 0.5
x_values = [0, 1]
pdf_values = [bernoulli_pdf(p, x) for x in x_values]
print("Bernoulli PDF:", pdf_values)
```

```python
#9
def calculate_z_scores(data):
    return (data - np.mean(data)) / np.std(data)

data = np.random.normal(50, 10, 100)
z_scores = calculate_z_scores(data)

plt.hist(z_scores, bins=30, density=True, alpha=0.7, edgecolor='black')
plt.xlabel("Z-Score")
plt.ylabel("Density")
plt.title("Z-Score Distribution")
plt.show()
```

```python
#14
data_point = 55
mean = 50
std_dev = 10
z_score = (data_point - mean) / std_dev
print("Z-Score:", z_score)
```

```python
#15
from scipy.stats import norm

sample_mean = 52
population_mean = 50
std_dev = 10
n = 30

z_statistic = (sample_mean - population_mean) / (std_dev / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z_statistic)))

print("Z-Statistic:", z_statistic)
print("P-Value:", p_value)
```

```python
#16
sample = np.random.normal(50, 10, 30)
mean, std_dev = np.mean(sample), np.std(sample, ddof=1)
confidence = 0.95
margin_error = stats.norm.ppf((1 + confidence) / 2) * (std_dev / np.sqrt(len(sample)))

ci_lower, ci_upper = mean - margin_error, mean + margin_error
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```
